```python
#Import Dependencies 
import pandas as pd
import csv
from sqlalchemy import create_engine
import psycopg2
```

```python
#Import CSV files & Cleaning our data
csv_path= ("data/201901-citibike-tripdata.csv") 
citibike119= pd.read_csv(csv_path)
citibike119.head()
```

```python
# Drop Null values
citibike119 = citibike119.dropna(how='any', axis=0)
```

```python
# Count DataFrame records to make sure no null values
citibike119.count()
```

```python
# Change Dataframe column names to match SQL table column name
citibike119.columns = ['trip_duration', 'start_time', 'stop_time', 'start_station_id', 'start_station_name', 'start_lat', 
                       'start_long','end_station_id', 'end_station_name', 'end_lat', 'end_long', 'bike_id', 'user_type',
                      'birth_year', 'gender']
```

```python
# Connect to Postgressql 
connection_string = "postgres:Reham@77@localhost:5432/citibike_tripdata"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
# Load data to SQL table
citibike119.to_sql(name='citibike119', con=engine, if_exists='append', index=False)
```
